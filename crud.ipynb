# DELETAR

# CONSULTA COM JOIN

```python
cursor.execute("select consulta.dt_dia 'dia_consulta', paciente.p_nome 'nome_paciente', medico.p_nome 'nome_medico', medico.crm 'crm', medico.especialidade 'especialidade', consulta.resultado_exame 'resultado_exame', consulta.receita 'receita' from consulta inner join paciente on consulta.fk_cpf_paciente = paciente.cpf_p inner join medico on consulta.fk_crm_medico = medico.crm")
rows = cursor.fetchall()
df = pd.DataFrame.from_records(rows, columns=[col[0] for col in cursor.description])
df
```
